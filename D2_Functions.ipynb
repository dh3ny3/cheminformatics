```python
"""
DAY 02: Functions as SOPs
==================================================
SYNTAX:
    def function_name(param1, param2):
        if validation_check:
            return error_message
        result = calculation
        return round(result, decimals)

CHEMISTRY:
    - Molarity: moles(mol) / volume(L); concentration unit (mol/L)
    - Significant Figures: Round to match lab instrument precision
    - Unit Consistency: Never mix g/mg or L/mL without conversion
"""
```

```python
#defining functions: SOP
def calculate_molarity(moles, volume_litres):
    """
    Calculate Molarity (concentration)

    Parameters:
    Moles: Amount of substance (in grams)
    Volume: Litres of liquid the substance is dissolved in

    Returns:
    Concentration in mol/L
    """
    if volume_litres == 0:
        return "Error: Division by Zero"
    molarity = moles / volume_litres
    return round(molarity, 4)

#example
result_01 = calculate_molarity(0.5, 0.25) #2.0M
result_02 = calculate_molarity(0.1, 0.5) #0.2M
result_03 = calculate_molarity(1.0, 0) #error case

print(f"Solution 1: {result_01} M")
print(f"Solution 2: {result_02} M")
print(f"Solution 3: {result_03}")
```
