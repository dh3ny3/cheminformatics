```python
#list of dictionaries (industry standard)
compounds = [
    {"name": "Aspirin",    "mw": 180.16, "logp": 1.19,  "hbd": 1, "hba": 4},
    {"name": "Caffeine",   "mw": 194.19, "logp": -0.07, "hbd": 0, "hba": 6},
    {"name": "Atorvastin", "mw": 558.64, "logp": 4.66,  "hbd": 2, "hba": 6}, #fails MW
    {"name": "Paclitaxel", "mw": 853.91, "logp": 3.00,  "hbd": 4, "hba": 14}, #fails MW and HBA
    {"name": "Metformin",  "mw": 129.16, "logp": -1.43, "hbd": 3, "hba": 1}
]

#storing sorted results
ro5_pass = []
ro5_fail = []

#loop for high-throughput screening
for cpd in compounds:
    passes = (
        cpd["mw"] <= 500 and
        cpd["logp"] <= 5 and
        cpd["hbd"] <= 5 and
        cpd["hba"] <= 10
    )

    if passes:
        ro5_pass.append(cpd["name"])
    else:
        reasons = []
        if cpd["mw"] > 500: reasons.append(f"MW={cpd['mw']}")
        if cpd["logp"] > 5: reasons.append(f"log P={cpd['logp']}")
        if cpd["hbd"] > 5: reasons.append(f"HBD={cpd['hbd']}")
        if cpd["hba"] > 10: reasons.append(f"HBA={cpd['hba']}")
        ro5_fail.append({"name": cpd["name"], "failed": reasons})

print(f"Oral Drug Candidates ({len(ro5_pass)}):\n{ro5_pass}")
print(f"\nRejected ({len(ro5_fail)}): ")
for item in ro5_fail:
    print(f"{item['name']}: Failed because {', '.join(item['failed'])}")
```
