```python
"""
Day 4: Molecular Property Lookup Tool with Dictionaries
=======================================================
SYNTAX:
    nested_dict = {"key": {"sub_key": value}}   # Structured database format
    dict.get(key, default)                      # Safe lookup (no KeyError)
    str.lower()                                 # Case normalization
    f"{var:<12}"                                # Fixed-width formatting

CHEMISTRY:
    - Drug Database Schema: Mirrors PubChem/ChEMBL JSON record structure
    - LogP: Partition coefficient; lipophilicity & membrane permeability predictor
    - HBD/HBA: Hydrogen Bond Donors/Acceptors; solubility & binding specificity markers
    - Oral Bioavailability: Fraction of dose reaching systemic circulation via GI tract
    - Therapeutic Class: Pharmacological categorization for drug repurposing
"""
```

```python
#building a list of databases
#each entry mirrors a PubChem/ChEMBL record structure
drug_database = {
    "Aspirin": {
        "SMILES": "CC(=O)Oc1ccccc1C(=O)O",
        "MW": 180.16,
        "log P": 1.19,
        "HBD": 1,
        "HBA": 4,
        "Oral Bioavailability": True,
        "Therapeutic Class": "NSAID"
    },
    "Caffeine": {
        "SMILES": "CN1C=NC2=C1C(=O)N(C(=O)N2C)C",
        "MW": 194.19, 
        "log P": -0.07,
        "HBD": 0,
        "HBA": 6,
        "Oral Bioavailability": True,
        "Therapeutic Class": "CNS Stimulant"
    },
    "Ibuprofen": {
        "SMILES": "CC(C)Cc1ccc(cc1)C(C)C(=O)O",
        "MW": 206.28,
        "log P": 3.97,
        "HBD": 1,
        "HBA": 2,
        "Oral Bioavailability": True,
        "Therapeutic Class": "NSAID"
    },
    "Paclitaxel": {
        "SMILES": "CC1=C2C(C(=O)C3(C(CC4C(C3C(C(C2(C)C)(CC1OC(=O)C(C(C5=CC=CC=C5)NC(=O)C6=CC=CC=C6)O)O)OC(=O)C7=CC=CC=C7)(CO4)OC(=O)C)O)C)OC(=O)C",
        "MW": 853.91,
        "log P": 3.00,
        "HBD": 4,
        "HBA": 14,
        "Oral Bioavailability": False,
        "Therapeutic Class": "Anticancer"
    },
    "Metformin": {
        "SMILES": "CN(C)C(=N)N=C(N)N",
        "MW": 129.16,
        "log P": -1.43,
        "HBD": 3,
        "HBA": 1,
        "Oral Bioavailability": True,
        "Therapeutic Class": "Antidiabetic"
    }
}

#lookup function
def lookup_drug(drug_name):
    """
    Retrieve all properties for a given drug.
    
    Parameters:
        drug_name (str): Case sensitive drug name
        
    Returns:
        dict: All stored properties; or an error message if not found
    """
    #convert input into lowercase
    search_key = drug_name.lower()

    #convert database keys into lowercase
    for key in drug_database:
        if key.lower() == search_key:
            return drug_database[key] #returns original case data
    return f" '{drug_name}' not found in database. Please check spelling or add entry."
    
#usage demonstration

print("=" * 50)
print("Molecular Lookup Directory Tool")
print("=" * 50)

#e.g., testing a known drug

aspirin_data = lookup_drug("Aspirin")
print(f"\n Aspirin:")
for prop, value in aspirin_data.items():
    print(f" {prop:>20}: {value}")

#test unknown drug
print(f"\n{lookup_drug('Paracetamol')}")

#batch comparison
print("\n Oral Drug Bioavailability:")
for name, props in drug_database.items():
    status = "Oral" if props["Oral Bioavailability"] else "IV Only"
    print(f" {name:<12} | MW={props['MW']:>6.2f} | LogP={props['log P']:>5.2f} | {status}")
```
